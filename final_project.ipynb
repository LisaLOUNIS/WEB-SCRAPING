## Scraping using BeautifulSoup

```python
import requests
from bs4 import BeautifulSoup

i = 1

for url in offer_urls:
    response = requests.get(url)
    data = response.text
    soup = BeautifulSoup(data, 'html.parser')
    print(i)
    title = soup.find('h2').get_text()
    company = soup.find('span', class_="wui-text").get_text()
    location = soup.find('span', class_='sc-1eoldvz-0 bZJPQK').get_text()
    profil_recherche_section = soup.findAll('div', class_='sc-18ygef-1 ezamTS')[1] 
    lists = profil_recherche_section.find_all('ul')
    competences = []
    for ul in lists:
        items = ul.find_all('li')
        for item in items:
            competences.append(item.get_text(strip=True))
    print(f"Job offer: {title}")
    print(f"Company: {company}")
    print(f"Location: {location}")
    print(f"Demanded skills:")
    for c in competences:
        print(f" - {c}")
    print("----------------------------------------------------------")
    i += 1
```

## Scraping using Selenium

```python
from selenium import webdriver
from selenium.webdriver.common.by import By

driver_path = '/path/to/chromedriver'

i = 1

for url in offer_urls:
    driver = webdriver.Chrome()
    driver.get(url)

    print(i)

    title = driver.find_element(By.TAG_NAME, 'h2').text
    company = driver.find_element(By.CLASS_NAME, 'wui-text').text
    location = driver.find_element(By.CLASS_NAME, 'sc-1eoldvz-0.bZJPQK').text

    profil_recherche_section = driver.find_elements(By.CLASS_NAME, 'sc-18ygef-1.ezamTS')[1]
    lists = profil_recherche_section.find_elements(By.TAG_NAME, 'ul')
    competences = []
    for ul in lists:
        items = ul.find_elements(By.TAG_NAME, 'li')
        for item in items:
            if item.text:
                competences.append(item.text)

    print(f"Job offer: {title}")
    print(f"Company: {company}")
    print(f"Location: {location}")
    print("Demanded skills:")
    for c in competences:
        print(f" - {c}")
    print("----------------------------------------------------------")
    i += 1
    driver.quit()
```

## Scraping top RSE companies

```python
import time
from selenium.webdriver.common.by import By

driver = webdriver.Chrome()

company_scores = []

for url in companies_urls:
    driver.get(url)
    time.sleep(1)
    company_elements = driver.find_elements(By.CLASS_NAME, 'company-item__inner')
    for company_element in company_elements:
        name = company_element.find_element(By.TAG_NAME, 'h3').text.strip()
        rank = company_element.find_element(By.CLASS_NAME, 'company-item__position').text.strip()
        score = company_element.find_element(By.CLASS_NAME, 'company-item__score').text.strip().split("\n")[0]
        company = {"company": name, "rank": rank, "score": score}
        company_scores.append(company)
        print(company)

driver.quit() 
```

## Collecting data scraped from Google API

```python
import os
import pandas as pd

folder_path = "google_api"
file_names = [f for f in os.listdir(folder_path) if f.endswith('.csv')]
file_names.sort()

combined_df = pd.DataFrame()

for i, file_name in enumerate(file_names):
    file_path = os.path.join(folder_path, file_name)

    if i == 0:
        combined_df = pd.read_csv(file_path)
    else:
        df = pd.read_csv(file_path, header=None, skiprows=1)
        df.columns = combined_df.columns
        combined_df = pd.concat([combined_df, df], ignore_index=True)

offers_df = combined_df
offers_df.dropna(inplace=True)
offers_df
```

```python
company_df = pd.DataFrame(company_scores)
company_df
```

```python
def find_partial_match(company_name, company_df):
    for index, row in company_df.iterrows():
        if company_name in row['company'] or row['company'] in company_name:
            return row['rank'], row['score']
    return None, None

offers_df['rank'], offers_df['score'] = zip(*offers_df['company_name'].apply(lambda x: find_partial_match(x, company_df)))
offers_df
```

```python

import ast

def parse_country(location):
    location_lower = location.lower()
    if "belgium" in location_lower:
        return "Belgium"
    elif "canada" in location_lower:
        return "Canada"
    elif "france" in location_lower:
        return "France"
    elif "germany" in location_lower:
        return "Germany"
    elif "ireland" in location_lower:
        return "Ireland"
    elif "spain" in location_lower:
        return "Spain"
    elif "switzerland" in location_lower:
        return "Switzerland"
    elif "japan" in location_lower:
        return "Japan"
    elif "portugal" in location_lower:
        return "Portugal"
    else:
        return location
    
def process_skills(skills):
    try:
        skills_list = ast.literal_eval(skills)  
        return list(set(skill.strip().lower() for skill in skills_list))
    except (ValueError, SyntaxError):
        return skills  

offers_df['technical_skills'] = offers_df['technical_skills'].apply(process_skills)
offers_df['technical_skills'] = offers_df['technical_skills'].apply(lambda skills: [skill.lower() for skill in skills])

offers_df
```
